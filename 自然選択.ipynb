```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import random 
```

0→a遺伝子　1→A遺伝子

```python
np.random.seed(seed=1)
N = 100 #世代
s = 0.1 #Aの適応度に加算
p = 0.1 #Aの頻度
w_A = 1+s
w_a = 1

#０世代作成
children_size = 100
elem = [0,1] #aを０、Aを1に対応
parents = np.random.choice(a=elem,size=children_size,p=[1-p,p])

freq_A = [sum(parents)/len(parents)]

for i in range(N-1):
    choosen_a_prob = (1-p)/(1+s*p)
    choosen_A_prob = (w_A)*p/(1+s*p)
    children = np.random.choice(a=elem,size=children_size,p=[choosen_a_prob,choosen_A_prob])
    p = sum(children) / len(children)
    freq_A.append(p)

plt.plot(range(N),freq_A)
plt.xlabel('世代')
plt.ylabel('Aの出現率')
plt.title('Aの進化のシミュレーション')
plt.grid()
plt.show()
```

```python
N = 100 #世代
s = 0.1 #Aの適応度に加算
p = 0.1 #Aの頻度
w_A = 1+s
w_a = 1

#０世代作成
children_size = 100
elem = [0,1] #aを０、Aを1に対応
parents = np.random.choice(a=elem,size=children_size,p=[1-p,p])

#Aの出現率を格納
freq_A = [sum(parents)/len(parents)]

#100回シミュレーション
for j in range(100):
    np.random.seed(seed=j)
    p = 0.1
    freq_A = [p]
    for i in range(N-1):
        choosen_a_prob = (1-p)/(1+s*p)
        choosen_A_prob = (w_A)*p/(1+s*p)
        children = np.random.choice(a=elem,size=children_size,p=[choosen_a_prob,choosen_A_prob])
        p = sum(children) / len(children)
        freq_A.append(p)
    plt.plot(range(N),freq_A,alpha=0.4,color='grey',zorder=0)

#予測値
p = 0.1
freq_A_expect = [p]
for i in range(N-1):
    p = (w_A)*p/(1+s*p)
    freq_A_expect.append(p)

plt.scatter(range(N),freq_A_expect,zorder=1)
plt.xlabel('世代')
plt.ylabel('Aの出現率')
plt.title('Aの進化のシミュレーション')
plt.grid()
plt.show()
```
