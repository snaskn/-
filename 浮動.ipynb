```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
import random
N = 30
i = 15
y = 1000
rd = 3
freqs = []
rdint = random.randint(1,100)

for j in range(rd):
    np.random.seed(rdint+j)
    freq = []
    arr = np.concatenate((np.ones(N//2), np.zeros(N-N//2)))
    for _ in range(y):
        i = sum(arr)
        p_i = i / N
        freq.append(p_i)
        rm = np.random.choice([1,0],p=[p_i,1-p_i])
        arr = np.delete(arr,rm)
        ap = np.random.choice([1,0],p=[p_i,1-p_i])
        arr = np.append(arr,ap)
    freqs.append(freq)

for j in range(rd):
    plt.plot(range(1,y+1),freqs[j])

plt.title('モランモデルのシミュレーション')
plt.xlabel('年代')
plt.ylabel('Aの頻度')
plt.ylim(-0.05,1.05)
plt.grid()
plt.savefig('moran model simulation')
```
